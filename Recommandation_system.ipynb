# Content Based Recommendation System

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("asaniczka/tmdb-movies-dataset-2023-930k-movies")

print("Path to dataset files:", path)
```

```python
import os

files = os.listdir(path)
data_dir = os.path.join(path, files[0])
```

```python
import pandas as pd

df = pd.read_csv(data_dir)
```

```python
df.head()
```

```python
df.isnull().sum()
df.duplicated().sum()
df.drop_duplicates()
df = df.dropna()
```

```python
new_df = df[['id', 'title', 'overview', 'genres', 'tagline', 'keywords']]
```

```python
new_df.head()
```

```python
import string

def transform_text(text):
  text = text.lower()
  return text.replace(' ', '')
```

```python
new_df["genres"] = new_df["genres"].apply(transform_text)
new_df["keywords"] = new_df["keywords"].apply(transform_text)
```

```python
new_df.head()
```

```python
new_df["text"] = df["overview"] + df["tagline"] + df["genres"] + df["keywords"]
```

```python
new_df.head()
```

```python
new_df = new_df[["id", "title", "text"]]
```

```python
new_df["text"] = new_df["text"].apply(lambda x: x.lower())
new_df["text"] = new_df["text"].apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))

from nltk.stem import PorterStemmer
ps = PorterStemmer()
new_df["text"] = new_df["text"].apply(lambda x: ' '.join([ps.stem(word) for word in x.split()]))
```

```python
new_df["title"] = new_df["title"].apply(lambda x: x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=10000, stop_words="english")
vector = cv.fit_transform(new_df["text"]).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vector)
```

```python
new_df[new_df['title'] == "iron man"].index
```

```python
def recommended(movie):

  movie = movie.lower()
  movie_row = new_df[new_df["title"] == movie]
  if movie_row.empty:
    print(f"Movie '{movie.title()}' not found in the database. Please check the title.")
    return

  movie_index = movie_row.index[0]
  distance = similarity[movie_index] # Corrected from similarity(movie_index)
  movies_list = sorted(list(enumerate(distance)), reverse=True, key=lambda x: x[1])[1:10]

  print(f"Recommendations for '{movie.title()}':")
  for i in movies_list:
    print(new_df.iloc[i[0]].title.title())
```

```python
user_input = input("Enter a movie title: ")
recommended(user_input)
```

```python
import requests

def fetch_poster(movie_id):
    url = f"https://api.themoviedb.org/3/movie/{movie_id}?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US"
    try:
        data = requests.get(url)
        data.raise_for_status() # Raise HTTPError for bad responses (4xx or 5xx)
        data = data.json()
        poster_path = data.get('poster_path')
        if poster_path:
            full_path = "https://image.tmdb.org/t/p/w500/" + poster_path
            return full_path
        else:
            print(f"Poster path not found for movie ID {movie_id}.")
            return None
    except requests.exceptions.HTTPError as e:
        print(f"HTTP error for movie ID {movie_id}: {e.response.status_code} - {e.response.text}")
        return None
    except requests.exceptions.ConnectionError as e:
        print(f"Connection error for movie ID {movie_id}: {e}")
        return None
    except requests.exceptions.Timeout as e:
        print(f"Timeout error for movie ID {movie_id}: {e}")
        return None
    except requests.exceptions.RequestException as e:
        print(f"An unexpected error occurred for movie ID {movie_id}: {e}")
        return None
    except Exception as e:
        print(f"An unexpected error occurred during JSON parsing or data access for movie ID {movie_id}: {e}")
        return None
```
